```python
import pandas as pd
data = pd.read_csv('./bcodmo_dataset_3106_712b_5843_9069.csv')
data = data.drop(data.index[0])
data = data.reset_index(drop=True)

# time stamp 생성
data['date'] = pd.to_datetime(data['year'].astype(str) + '-' + data['month_local'].astype(str) + '-' + data['day_local'].astype(str), format='%Y-%m-%d')
data['date']

data
```

```python
# haul_id와 위도 , 경도와의 산관 관계 파악
unique_locations = data.groupby('haul_id')[['latitude', 'longitude']].first().reset_index()
unique_locations
```

```python
# haul_id와 경도 위도 의 중복값 확인
unique_counts = data.groupby('haul_id').apply(lambda x: x[['latitude', 'longitude']].drop_duplicates().shape[0])
duplicated_haul_ids = unique_counts[unique_counts > 1]

if duplicated_haul_ids.empty:
    print("중복된 위치 값을 가진 haul_id는 없습니다.")
else:
    print(f"다음의 haul_id에 중복된 위치 값이 있습니다:\n{duplicated_haul_ids}")
```

```python
# dtype 변경
for col in ['length_froz' , 'length_wet' , 'weight_froz' , 'weight_wet' , 'stomwt_full' , 'stomwt_empty']:
    data[col] = data[col].astype(float)
```

```python
# groupby를 이용해서 데이터 전처리
# 각 개체에 대해서 데이터를 처리해놓은게 아니라 그물에서 나오는 생선을 다 기록한것 같음 
grouped_data = data.groupby(['date', 'haul_id', 'species_common_name', 'maturity'])[['weight_froz', 'weight_wet', 'length_froz', 'length_wet', 'stomwt_full', 'stomwt_empty']].mean().reset_index()
grouped_data
```

```python
# 위 데이터를 그물 , 종 , 성숙도를 기준으로 시각화

import plotly.express as px

grouped_data = data.groupby(['haul_id', 'species_common_name', 'maturity', 'date'])['weight_wet'].mean().reset_index()

fig = px.line(grouped_data, 
              x='date', 
              y='weight_wet', 
              color='species_common_name', 
              line_dash='maturity', 
              title='Weight_wet over Time by Species and Maturity',
              labels={'weight_wet': 'Average Weight_wet', 'date': 'Date'},
              )
fig.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns  # Seaborn is also utilized for enhanced visualization

# 'haul_id', 'species_common_name', 'maturity', 'date'를 기준으로 그룹화하고 
# 'weight_wet'의 평균값을 계산
grouped_data = data.groupby(['haul_id', 'species_common_name', 'maturity', 'date'])['weight_wet'].mean().reset_index()

# Unique species and maturities for legends and line styles
species = grouped_data['species_common_name'].unique()
maturities = grouped_data['maturity'].unique()

# Create a colormap for the species
colors = sns.color_palette('tab10', len(species))
color_map = dict(zip(species, colors))

# Define line styles for different maturities
line_styles = ['-', '--', '-.', ':']
line_map = dict(zip(maturities, line_styles))

plt.figure(figsize=(14, 8))

# Loop and plot
for specie in species:
    for maturity in maturities:
        subset = grouped_data[(grouped_data['species_common_name'] == specie) & (grouped_data['maturity'] == maturity)]
        if not subset.empty:
            plt.plot(subset['date'], subset['weight_wet'], 
                     label=f"{specie} ({maturity})", 
                     color=color_map[specie], 
                     linestyle=line_map[maturity])

plt.title('Weight_wet over Time by Species and Maturity')
plt.xlabel('Date')
plt.ylabel('Average Weight_wet')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```
