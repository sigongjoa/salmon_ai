### preprocessing 

이 데이터도 개체별로 데이터를 측정한것이 아니라서 `시간` , `생육상태` 기준으로 groupby의 mean으로 해서 평균 데이터를 사용해서 데이터를 merge

```python
feed_data
```

```python
feed_data['공급일자'] = pd.to_datetime(feed_data['공급일자'], format='%Y%m%d')
feed_data.rename(columns={'공급일자': 'time'}, inplace=True)
feed_data
```

```python
group_feed_data = feed_data.groupby(['수조번호', 'time','생육상태']).mean()
group_feed_data.reset_index(inplace=True)
group_feed_data
```

```python
water_data = pd.read_csv('연어 수질분석데이터.csv')
water_data = water_data[water_data.columns[:10]]
water_data
```

```python
water_data['측정일자'] = pd.to_datetime(water_data['측정일자'], format='%Y%m%d')
water_data.rename(columns={'측정일자': 'time'}, inplace=True)
water_data
```

```python
group_water_data = water_data.groupby(['수조번호', 'time','생육상태']).mean()
group_water_data.reset_index(inplace=True)
group_water_data
```

* 결측치 처리가 애매해서 megre 이후에 날릴 예정 

```python
sensor_data = pd.read_csv('센서데이터_연어.csv')
sensor_data
```

```python
import pandas as pd

sensor_data['date'] = pd.to_datetime(sensor_data['date']).dt.date
sensor_data['date'] = pd.to_datetime(sensor_data['date'] , format='%Y-%m-%d')
sensor_data.rename(columns={'date': 'time'}, inplace=True)
sensor_data.rename(columns={'수조': '수조번호'}, inplace=True)
sensor_data.rename(columns={'구분': '생육상태'}, inplace=True)


grouped_sensor_data = sensor_data.groupby(['time', '수조번호', '생육상태']).mean().reset_index()
grouped_sensor_data
```

```python
grouped_sensor_data[grouped_sensor_data['FLOW(L)'].isnull()]
```

```python
grouped_sensor_data[grouped_sensor_data['Temp_air'].isnull()]
```

```python
die_data = pd.read_csv('연어 폐사관리데이터.csv')
die_data
```

```python
die_data['측정일자'] = pd.to_datetime(die_data['측정일자'], format='%Y%m%d')
die_data.rename(columns={'측정일자': 'time'}, inplace=True)
die_data
```

```python
group_die_data = die_data.groupby(['수조번호', 'time','생육상태']).mean()
group_die_data.reset_index(inplace=True)
group_die_data
```

```python
target_data = pd.read_csv('연어 샘플링 실측데이터(길이,무게).csv')
target_data
```

```python
target_data['측정일자'] = pd.to_datetime(target_data['측정일자'], format='%Y%m%d')
target_data.rename(columns={'측정일자': 'time'}, inplace=True)
target_data
```

```python
group_target_data = target_data.groupby(['수조번호', 'time','생육상태']).mean()
group_target_data.reset_index(inplace=True)
group_target_data
```
