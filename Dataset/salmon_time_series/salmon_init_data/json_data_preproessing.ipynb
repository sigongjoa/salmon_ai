```python
import os
import json
import pandas as pd

base_dir = '../../salmon_segmentation/out/2.라벨링데이터'
json_files = []

for subdir in os.listdir(base_dir):
    subdir_path = os.path.join(base_dir, subdir)
    
    if os.path.isdir(subdir_path):        
        for filename in os.listdir(subdir_path):
            if filename.endswith('.json'):
                json_files.append(os.path.join(subdir_path, filename))


```

```python
for json_file in json_files:
    print(json_file)
```

```python
# json file print
for json_file in json_files:
    with open(json_file, 'r', encoding='utf-8') as f:
        data = json.load(f)
        print(f"Contents of {json_file}:\n")
        print(json.dumps(data, indent=4, ensure_ascii=False))
        print("-" * 50)
    break
```

```python
# json data to dataframe
df = pd.json_normalize(data)
df
```

```python
# 전체 데이터 append
all_data = pd.DataFrame()
for json_file in json_files:
    with open(json_file, 'r', encoding='utf-8') as f:
        data = json.load(f)
        df = pd.json_normalize(data)
        all_data = all_data.append(df, ignore_index=True)
all_data
```

```python
# feed 데이터 처리 dict를 column으로 가져옴
all_data['growth_management.feed_'] = all_data['growth_management.feed'].apply(lambda x: x[0] if isinstance(x, list) and x else {})

feed_df = pd.json_normalize(all_data['growth_management.feed_'])
feed_df.columns = ['feed1_' + str(col) for col in feed_df.columns]
all_data = pd.concat([all_data, feed_df], axis=1)

all_data.drop('growth_management.feed_', axis=1, inplace=True)

all_data.head()
```

```python
all_data['growth_management.feed_'] = all_data['growth_management.feed'].apply(lambda x: x[1] if isinstance(x, list) and x else {})

feed_df = pd.json_normalize(all_data['growth_management.feed_'])
feed_df.columns = ['feed2_' + str(col) for col in feed_df.columns]
all_data = pd.concat([all_data, feed_df], axis=1)

all_data.drop('growth_management.feed_', axis=1, inplace=True)

all_data.head()
```

```python
all_data.columns
```

```python
# 급이 데이터에서 %로 되어있는 것을 실제 gram으로 변경
all_data['feed1_nutrition.protein'] = all_data['feed1_nutrition.protein'].str.rstrip('%').astype('float') / 100.0
all_data['feed2_nutrition.protein'] = all_data['feed2_nutrition.protein'].str.rstrip('%').astype('float') / 100.0

all_data['feed1_nutrition.fat'] = all_data['feed1_nutrition.fat'].str.rstrip('%').astype('float') / 100.0
all_data['feed2_nutrition.fat'] = all_data['feed2_nutrition.fat'].str.rstrip('%').astype('float') / 100.0

all_data['fee1_protein'] = all_data['feed1_nutrition.protein'] * all_data['feed1_amount.value']
all_data['fee2_protein'] = all_data['feed2_nutrition.protein'] * all_data['feed2_amount.value']

all_data['fee1_fat'] = all_data['feed1_nutrition.fat'] * all_data['feed1_amount.value']
all_data['fee2_fat'] = all_data['feed2_nutrition.fat'] * all_data['feed2_amount.value']
```

```python
# time stamp 처리 : 일 단위로 변경
all_data['growth_management.water.MeasuredAt'] = pd.to_datetime(all_data['growth_management.water.MeasuredAt'])
all_data['growth_management.water.MeasuredAt'] = all_data['growth_management.water.MeasuredAt'].dt.date
all_data['growth_management.water.MeasuredAt'].value_counts()
```

```python
# 실제 사용하는 데이터만 가져옴 
cleaned_data = all_data[['category.name' , 'category.age.min' , 'category.age.max' , 'tank.id' , 
                         'growth_management.water.ammonium.value' , 'growth_management.water.nitrite.value' , 
                         'growth_management.water.nitrate.value' , 'growth_management.water.MeasuredAt' , 
                         'water.temperature.value' , 'water.salinity.value' , 'water.do.value' , 'water.ph.value' , 'water.orp.value',
                          'water.illuminance.value' , 'fee1_protein' , 'fee2_protein' , 'fee1_fat' , 'fee2_fat']]
cleaned_data
```
