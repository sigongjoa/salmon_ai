```python
from ftplib import FTP
import os

def download_from_ftp(host, path, local_path='.'):
    with FTP(host) as ftp:
        ftp.login()  # If authentication is required, provide username and password here.
        ftp.cwd(path)
        filenames = ftp.nlst()  # List of filenames in the current directory
        
        if not os.path.exists(local_path):
            os.makedirs(local_path)

        for filename in filenames:
            with open(os.path.join(local_path, filename), 'wb') as local_file:
                ftp.retrbinary('RETR ' + filename, local_file.write)

# Usage
download_from_ftp('ftp.nefsc.noaa.gov', '/pub/dropoff/PARR/READ/PSB/NEST/27493/', 'local_directory_path')
```
